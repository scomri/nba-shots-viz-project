# Preproccess for Viz #3 - Basketball Court Rim Locations

```python
df = pd.read_csv('nba_all_clean.csv')
```

```python
def data_nba_and_rim(data):
    data_all_df = data
    # center_of_rim = [-25, +2.65]
    data_all_df['shotX_rim'] = (data_all_df['shotX'] - 25) * 10
    data_all_df['shotY_rim'] = (data_all_df['shotY'] - 2.65) * 10
    return data_all_df

df = data_nba_and_rim(df)
```

```python
df = df[['shotX_rim', 'shotY_rim', 'shot_type', 'made', 'distance', 'team']]
```

```python
# Load the data frame
# Separate the data frame based on shot_type
df_shot_type_2 = df[df['shot_type'] == 2].copy()
df_shot_type_3 = df[df['shot_type'] == 3].copy()
```

```python
df_shot_type_2_made_True = df_shot_type_2[df_shot_type_2['made'] == True]
print(df_shot_type_2_made_True.shape[0])

df_shot_type_2_made_False = df_shot_type_2[df_shot_type_2['made'] == False]
print(df_shot_type_2_made_False.shape[0])

df_shot_type_3_made_True = df_shot_type_3[df_shot_type_3['made'] == True]
print(df_shot_type_3_made_True.shape[0])

df_shot_type_3_made_False = df_shot_type_3[df_shot_type_3['made'] == False]
print(df_shot_type_3_made_False.shape[0])
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from tqdm import tqdm

dfs = [df_shot_type_2_made_True, df_shot_type_2_made_False, df_shot_type_3_made_True, df_shot_type_3_made_False]
columns_to_cluster = ['shotX_rim', 'shotY_rim']
clustered_dfs = []

for df in tqdm(dfs):
  k_shot_type_made = 300
  kmeans_shot_type_made = KMeans(n_clusters=k_shot_type_made, random_state=42)
  labels_shot_type_made = kmeans_shot_type_made.fit_predict(df[columns_to_cluster])
  df['cluster_label'] = labels_shot_type_made
  cluster_centers_shot_type_made = kmeans_shot_type_made.cluster_centers_
  curr_df = pd.DataFrame(cluster_centers_shot_type_made, columns=columns_to_cluster)
  clustered_dfs.append(curr_df)
```

```python
def pre(df, v1, v2):
  df.insert(1, 'made', v1)
  df.insert(1, 'shot_type', v2)
  return df

df0 = pre(clustered_dfs[0], True, 2)
df1 = pre(clustered_dfs[1], False, 2)
df2 = pre(clustered_dfs[2], True, 3)
df3 = pre(clustered_dfs[3], False, 3)

df = pd.concat([df0, df1, df2, df3], axis=0)
```

```python
df.to_csv('data_for_viz2.csv')
```
